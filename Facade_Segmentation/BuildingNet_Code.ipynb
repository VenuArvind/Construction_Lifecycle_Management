```python
import open3d as o3d
import numpy as np

# Load the .npy file
data = np.load('/content/drive/MyDrive/BuildingNet_Code/Output/Edge/adjacency/0.0001/node+minkownormal/checkpoint_best_acc/RESIDENTIALhouse_mesh5041.npy', allow_pickle=True)

# Convert data type to float64 and reshape if necessary
data = np.array(data, dtype=np.float64)

# Ensure data is 2D
if len(data.shape) == 1:
    data = data.reshape(-1, 1)

# Create a PointCloud object
point_cloud = o3d.geometry.PointCloud()

# Set the points of the point cloud to the loaded data
point_cloud.points = o3d.utility.Vector3dVector(data)

# Visualize the point cloud
o3d.visualization.draw_geometries([point_cloud])
```
